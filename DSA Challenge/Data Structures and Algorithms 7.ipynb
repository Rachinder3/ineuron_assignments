# 2. What are the qualities of a binary tree?

A binary tree is a hierarchial data structure in which each node has atmost 2 children. The child nodes are called the left child and the right child.

A binary tree node has the following 3 things:

1) Data Element

2) Pointer to store reference of left child.

3) Pointer to store reference of right child.

/////////////////////////////////////////////////////////


Some qualities of a binary tree are:

1) A binary tree can have maximum of 2^l nodes at level l, if the level of the root is zero.

////////////////////////////////////////////////////////

2) There exists a maximum of (2^h - 1) nodes in a binary tree if its height is h and the height of leaf node is one.

////////////////////////////////////////////////////////

3) If there exist L leaf nodes in a binary tree then it has atleast L+1 levels.

////////////////////////////////////////////////////////

4) A binary tree of n nodes has log2(n+1) minimum number of levels or minimum height.

////////////////////////////////////////////////////////

5) The minimum and maximum height of a binary tree having n nodes are ceil(log2 n) and n, respectively

////////////////////////////////////////////////////////

6) A binary tree of n nodes has (n+1) null references.



# 3. What is the best way to combine two balanced binary search trees?

This can be done by exlploiting 2 facts:
    
1) Inorder Traversal of a BST gives a sorted collection.

2) Merge functionality of Merge Sort Algorithm merges 2 sorted arrays.


///////////////////////////////////////////////////////////////////


The pseudocode for solving the given problem can be:


1) Do inorder traversal of the first tree, while traversing, store the elements in a list.

2) Do inorder traversal of the second tree, while traversing, store the elements in a list.


Now we have 2 sorted lists. We perform merge functionality of merge sort to combine the 2 sorted arrays.

3) Merge the 2 generated sorted arrays.



4) Generate the BST from the merged sorted array.

    

### Different codes

#### Inorder Traversal and storing in a list.

```python
def inorder(root, sorted_list):
    if root==None:
        return
    
    inorder(root.left) # recursively calling on the left subtree
    sorted_list.append(root.data) # storing the element in the sorted list
    inorder(root.right) # recursively calling on the right subtree
```

### Merging 2 sorted lists and returning the combined sorted list

```python
def merge(arr1,arr2):
    sorted_list = []
    
    i = 0
    j = 0 
    
    size_arr1 = len(arr1)
    
    size_arr2= len(arr2)
    
    k = 0
    
    while i<size_arr1 and j<size_arr2:
        if arr1[i] <= arr2[j]:
            sorted_list[k] = arr1[i] 
            i+=1
        else:
            sorted_list[k] = arr2[j]
            j+=1
        
        k+=1
    
    
    for l in range(i,size_arr1+1): # Adding the left over elements, if any
        sorted_list[k] = arr1[l]
        k+=1
        
    for l in range(j, size_arr2+1): # Adding the left over elements, if any
        sorted_list[k] = arr2[l]
        k+=1
        
    return sorted_list
        
        
    
```

### BST node

```python
class BST_Node():
    
    def __init__(self,data):
        self.data = data 
        self.right = None
        self.left = None
```

### Creating BST from 2 sorted lists

```python
def create_BST(arr, si, ei):
    
    if si > ei:
        return None  # return None if no element left in the lists
    
    
    midpt = (si+ei) / 2 # getting the midpt of the list
    
    element_at_mid = arr[midpt] # getting the element at the midpt
    
    newNode = BST_Node(element_at_mid) # creating a new node with element at mid
    
    newNode.left  = create_BST(arr, si, midpt-1) # calling recursively and storing the generated subtree as right child
    newNode.right = create_BST(arr, midpt+1, ei) # calling recursively and storing the generated subtree as left child
    
    return(newNode)
```

# 4. How would you describe Heap in detail?

Heap is a special kind of Binary Tree. 

A Heap is a:

1) Complete Binary Tree

2) Satisfies the Heap Order Property

Let's discuss these one by one.

////////////////////////////////////////////////////////

Complete Binary Tree:

A complete binary tree is a binary tree which is completely filled except the last level. The last level should be filled from left and right. Because of this property, heaps can be stored in the form of arrays.






##### Examples:

Here, all the levels are filled and the last level is filled from left to right

//////////////////////////////////////////////////////////////

Heap Order Property:

Heap Order propertry defines how elements are stored in a heap.

We can have 2 types of heap order properties and all the heaps should follow atleast 1 Heap Order Property.

1) Any given node is always greater than its child nodes and the key of the root node is the largest among all other nodes. 
The Heap that follows this property is called as <b>Max Heap</b>.

For Max Heap, for every node:  A[parent] >= A[children],
where A is the array storing the heap.


2) Any given node is always smaller than its child nodes and the key of the root node is smallest among all other nodes.
The Heap that follows this property us called as <b>Min Heap</b>.

For Min Heap, for every node: A[parent] <= A[children], where
A is the array storing the heap.


/////////////////////////////////////////////////////////

Application of Heaps:

1) Heaps are used to implement Priority Queues.

2) Heaps are used to implement Heapsort sorting algorithm.

///////////////////////////////////////////////////

We hash the object, using any available industry grade uniform hash function. Then we can mod this with the size of the underlying array which is storing the hashmap. With this we get the index.

//////////////////////////////////////////////////

# 7. How do you recursively sort a stack?

### Idea 

The idea is, we get the top of the stack. We call the sort stack function recursively and assume rest of the stack is now sorted. To place the stored top in its correct position, we use an additional stack, we remove from the top of the stack and place the elements in a second stack till the time the stored top is lesser than the current top. We place the stored top inside the stack. Then we get all the elements from the second stack and push them back into the original stack

### Pseudocode:

1) We run a recursive code.

2) Base Case: Number of elements <=1, stack is already sorted.

3) Remove the top of the stack and store it.

4) Recursively call the algorithm on the rest of the stack.

5) Once we get the control back, we use additional stack. 
Till stored element is less than top of stack, We pop from the original stack and push into the second stack.

6) Place the stored element into the original. 

7) Pop from the second stack and push the elements back into original stack.

### Implementation

```python
class stack:
    def __init__(self):
        self.__stack = [] # list that stores the elements of the stack
        self.__num_of_elements = 0 # stores the number of elements in the stack
        
    def push(self,data):
        self.__stack.append(data)
        self.__num_of_elements +=1
        
        
    def __str__(self): # overloads the print function
        return(" ".join([str(i) for i in self.__stack]))
    
    def pop(self):
        if self.__num_of_elements == 0:
            return -1 # avoiding underflow
        
        element = self.__stack.pop(self.__num_of_elements-1)
        self.__num_of_elements-=1
        
        return(element)
        
        
    def top(self):
        if self.__num_of_elements==0:
            return -1 # No element present in the stack
        return self.__stack[self.__num_of_elements-1]
    
    def num_elements(self):
        return self.__num_of_elements
    
    def sort_stack(self): # sorts the stack recursively
        
        if self.num_elements()<=1: # base case
            return 
        
       
        top_stack = self.pop() # storing the top of the stack
       
        
        self.sort_stack() # recursively calling the sort_stack function
        
        second_st = stack() # creating the second stack that stores the elements popped from original stack
        
        while top_stack < self.top(): # remove from the first stack, till the stored top remains lesser than current stack 
            #top and push into the second stack
            second_st.push(self.pop())
            
            
            
        self.push(top_stack)  # place the element in its right position in the current stack
       
        
        while second_st.num_elements() !=0: # get all the elements back from second stack and push them in the original stack
            self.push(second_st.pop())
            
            
            
        
        
    
            
```
