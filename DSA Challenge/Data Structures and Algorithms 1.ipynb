# 1. Describe Python's built-in data structure?

Python has 4 basic inbuilt data structures.

They are:

1) Lists

2) Dictionary

3) Tuple

4) Set

/////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

Lists: Lists store ordered collection of objects i.e. we can store a sequence of objects in a list.


Lists are mutable in nature unlike strings. Item reassignment is possible in lists.

/////////////////////////////////////////////////////////////////////////////////////////////////////////////////////


Dictionaries: Dictionaries are associate arrays. In standard arrays, the indexes can only be integers. In Dictionaries, indexes can be any object which is hashable.

So, in essence, Dictionaries store a collection of key value pairs. Keys of a dictionary are always unique.

This is like unordered hashmap in Cpp.


/////////////////////////////////////////////////////////////////////////////////////////////////////////////////////


Tuples: Tuples work exactly the same as lists except they are immuatble in nature. Item reassignment is not possible inside tuples. Since tuples are immutable, their length is fixed. To grow or shrink a tuple, a new tuple must be created.


/////////////////////////////////////////////////////////////////////////////////////////////////////////////////////


Sets: Sets are unordered collection of unique objects. 

Sets are unordered and we cannot access a set's element by using indexes.

# 2. Describe the Python user data structure?

User Defined data structures are those data structures which are not inbuilt in python but we can program them ourselves to reflect their functionality using concepts supported by python.

Some examples of user defined data structures are:

1) Linked List

2) Stack

3) Queue

4) Tree

5) Graph

////////////////////////////////////////////////////////////////////////////////////


Linked List : Linked List is a data structure used to store a collection of homogenous or hetrogenous data. Unlike arrays which store all the data in a contiguous block of memory, linked lists store data by storing data randomly in small blocks of memory. Each block is connected with another block either via pointers(C++) or via storing reference of one block in its parent's block(Python, Java etc).


////////////////////////////////////////////////////////////////////////////////////

Stack: Stack is a linear LIFO  data structure. LIFO stands for Last In, First Out.In a stack, the element that is inserted last, is removed first. Conversly, the element that is inserted first, is removed first.

Insertion operation is known as push() and deletion operation is known as pop()

////////////////////////////////////////////////////////////////////////////////////

Queue: Queue is a linear data structure that allows insertion of elements from one end and deletion from the other. Thus it follows, First In First Out(FIFO) methodology. The end which allows deletion is known as the front of the queue and the other end is known as the rear of queue.

Insertion operation into the queue is called as enqueue, Deletion operation from the queue is called dequeue.

////////////////////////////////////////////////////////////////////////////////////

Tree: A Tree is a non - linear, hierarchial data structure. 

Each node of a tree contains the data and the reference to its children nodes. 

The topmost node is called root. The nodes that don't have any children are called leaves. 

Trees are appropriate for storing data that isn't linearly connected but is hierarchial in nature.

////////////////////////////////////////////////////////////////////////////////////

Graph: A Graph is non linear data structure consisting of nodes and edges. The nodes are referred to as vertices and the edges are lines or arcs that connect any two nodes in the graph.


In essence a Graph is a finite set if vertices (nodes) and a set of edges which connect pair of nodes.


# 3. Describe the stages involved in writing an algorithm?

What is an algorithm??

An algorithm is a step by step set of instructions designed to solve a specific computer problem.

Before writing a program, a programmer must figure out the algorithm of the program. After that, the programmer should write the pseudocode. And only after that, write the actual program.

-----------------

The different stages involved in writing an algorithm are as follows:

1) Obtain detailed information on the problem:

Before starting with an algorithm, we must obtain maximum information about the problem that needs to be solved. 

2) Analyze the problem: 

Proper Analysis of the problem must be done including the data that must be gained, processed and retreived for generating a valid output.


3) Think of a problem solving approach:

The programmer has to come up with a problem solving approach that will help us to build the model to solve the given problem.

4) Review the problem solving approach and try to get a better alternative:

A programmer should always try and get the best possible and best optimized solution to any problem.  

Therefore, when we think of a problem solving approach, we should try thinking of a better alternative with better results. This will help us generate a finer programming logic.

Also, we should review the problem solving approach, to make sure the problem will be solved through it.


5) Develop a basic structure of Algorithm:

After the above steps are done, we should develop a basic structure of the problem solving approach. We should write down the approach step - by - step with short and effective description.


6) Optimize, Improve and refine:

After developing an Algorithm, we should try to optimize the explanation to increase its readability and accessibility.

Also, we should try to improve and refine the algorithm for better understanding and efficient for use.

# Outline the components of a good algorithm?

Some components of of a good algorithms are:

/////////////////////////////////////////////////
    
1) Input and Output must be specified: 
A good algorithm will always have well defined inputs and outputs, so that its easier for us to build the actual algorithm.

//////////////////////////////////////////////////

2) All important steps must be mentioned: 

An algorithm comprises of all short step-by-step processes that is performed in a program, thus every important step must be present in the Algorithm in considerate details.

The steps mentioned in an algorithm must lack grammatical mistakes as well so that misunderstanding and confusion can be avoided.

/////////////////////////////////////////////////

3) Instructions must be perfectly ordered:

The order of instructions is a very important. If we don't mantain the order of the instructions, our program won't work even if the logic is correct.

/////////////////////////////////////////////////

4) Short and Effective descriptions: 

The descriptions of the processes must be short, precise and effective so that reliability and efficiency of the program can be increased.

////////////////////////////////////////////////

5) Finite number of steps:

The algorithm should contain finite number of steps to generate a valid output. 

An Algorithm must conclude after performing certain operation and generating an output. The loops mentioned in the algorithm must terminate after performing the operations.

# 5. Describe the Tree traversal method?

Tree traversal methods help us in visting all the nodes in any Tree. 

Unlike linear data structures, Trees are hierarchial in nature. Hence, Trees can be traversed in many ways.

There are 4 main Tree Traversal methods:

1) Inorder

2) Preorder

3) Postorder

4) Levelorder

////////////////////////////////////////////////////////////////////

Let's discuss about each one with regards to Binary Tree:


1) Inorder Traversal:
This is a Depth First Search algorithm. Over here, we first visit the left subtree, then the root node and finally the right subtree.
We use stack for this traversal or we use recursion.

The order is: Left Root Right

////////////////////////////////////////////////////////////////////

2) PreOrder Traversal: 
This is a Depth First Search algorithm. Over here, we first visit the root, then the left subtree and finally the right subtree.
We use stack for this traversal or we use recursion.

The order is : Root Left Right

////////////////////////////////////////////////////////////////////

3) PostOrder Traversal:
This is once again a Depth First Search algorithm. Over here, we first visit the left subtree, then the right subtree and finally the root.
We use stack for this traversal or we use recursion.

The order is: Left Right Root

////////////////////////////////////////////////////////////////////

4) Level Order Traversal:
This is a Breadth First Search algorithm. We visit all the nodes in the same level before we move to the next level.
We uses queues for this traversal.
